## Gráfico de líneas

Este gráfico muestra cómo cambian los valores a lo largo del tiempo.  
Se utiliza para ver tendencias en datos continuos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = np.random.randint(10, 50, size=10)

plt.plot(x, y, marker='o')

plt.title("Gráfico de líneas")
plt.xlabel("Tiempo")
plt.ylabel("Valor")

plt.savefig("imgs/line_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Histograma

Este gráfico muestra la distribución de un conjunto de datos aleatorios.  
Se utiliza un histograma para ver cómo se agrupan los valores en diferentes rangos.  
Es útil para analizar la frecuencia de aparición de los datos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(100)

plt.hist(data, bins=15, color='purple', edgecolor='black')

plt.title("Histograma")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")

plt.savefig("imgs/histogram.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Scatter Plot

Este gráfico muestra la relación entre dos variables numéricas.  
Se utiliza un scatter plot para observar posibles correlaciones entre datos.  
Es útil para identificar patrones o agrupaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', label='Datos')

plt.title("Scatter Plot")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")

plt.legend()

plt.savefig("imgs/scatter_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Pie Chart

Este gráfico muestra la proporción de diferentes categorías dentro de un total.  
Se utiliza para representar porcentajes de forma clara.  
Es útil para visualizar cómo se divide un conjunto de datos.

```python
import matplotlib.pyplot as plt
```

```python

labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Pie Chart")

plt.savefig("imgs/pie_chart.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Box Plot

Este gráfico muestra la distribución de los datos y sus valores extremos.  
Se utiliza para analizar la dispersión y detectar posibles valores atípicos.  
Es útil en estadística descriptiva.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(100)

plt.boxplot(data)

plt.title("Box Plot")
plt.ylabel("Valores")

plt.savefig("imgs/box_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
